```python
def greatest(x,y,z):
    """greatest(x,y,z) ---> a greatest value
    it will look up all arguments as x,y,z and will return 
    greatest among x,y,z.""" # doc-string 
    if x >= y and x >= z : 
        return x
    elif y >= x and y >= z : 
        return y 
    else : 
        return z
    
print(greatest.__doc__)
result = greatest(10,30,20)
print(result)
```

```python
greatest(10)
```

```python
#formal arguments
def greatest(x,y=0,z=0):
    """greatest(x,y,z) ---> a greatest value
    it will look up all arguments as x,y,z and will return 
    greatest among x,y,z.""" # doc-string 
    if x >= y and x >= z : 
        return x
    elif y >= x and y >= z : 
        return y 
    else : 
        return z
    
print(greatest.__doc__)
result = greatest(10)
print(result)
```

```python
def poly(x,*constants) :
    """poly(x,*constants) --> will retrun solution of eqation 
    y = mx+c so on 
    """
    p = len(constants) - 1 
    s = 0
    for constant in constants : 
        s = s + ( x ** p )*constant
        p = p - 1
    return s   
    
```

```python
poly(5,1,2,3)
```

```python
import numpy as np 
```

```python
x_points = np.linspace(-50,50,20000)
```

```python
deg1 = (-16,3)
deg2 = (0.3,0.5,1.6)
deg3 = (0,1,5,1)
```

```python
deg1y = [ poly(x,*deg1) for x in x_points ] 
deg2y = [ poly(x,*deg2) for x in x_points ]
deg3y = [ poly(x,*deg3) for x in x_points ]
```

```python
import matplotlib.pyplot as plt

plt.plot(x_points,deg1y,'r')
plt.plot(x_points,deg2y,'g')
plt.plot(x_points,deg3y,'b')
plt.show()
```
